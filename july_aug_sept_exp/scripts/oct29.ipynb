```python
## asda

from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json

# Setup
results_dir = Path("/n/netscratch/dam_lab/Lab/drooryck/codeswitching-llms/results/oct30_moreplurals")
plots_base = Path("/n/home06/drooryck/codeswitching-llms/july_aug_sept_exp/scripts/temp_plots_oct30_moreplurals")
plots_base.mkdir(parents=True, exist_ok=True)

# Get all run directories
run_dirs = sorted([d for d in results_dir.iterdir() if d.is_dir() and d.name.startswith('p')])
print(f"Found {len(run_dirs)} run directories")

# Process each ablation type
for ablation_type in ['none', 'subject', 'verb', 'object']:
    print(f"\nProcessing ablation={ablation_type}...")
    
    orientation_data = []
    
    # Load orientation scores from JSON files
    for run_dir in run_dirs:
        parts = run_dir.name.split("_run")
        if len(parts) != 2:
            continue
        
        try:
            prop = float(parts[0][1:]) / 100.0
            run_id = int(parts[1])
        except ValueError:
            continue
        
        json_file = run_dir / f"ablation_{ablation_type}_metrics.json"
        if not json_file.exists():
            continue
        
        with open(json_file, 'r') as f:
            metrics = json.load(f)
        
        orientation_data.append({
            'prop': prop,
            'run_id': run_id,
            'fr_orientation': metrics['fr_orientation'],
            'nl_orientation': metrics['nl_orientation']
        })
    
    # Create DataFrame
    df = pd.DataFrame(orientation_data)
    print(f"  Loaded {len(df)} runs")
    
    # Create output directory
    plots_dir = plots_base / f"ablation_{ablation_type}"
    plots_dir.mkdir(parents=True, exist_ok=True)
    
    # Save CSV
    csv_path = plots_dir / 'orientation_scores.csv'
    df.to_csv(csv_path, index=False)
    print(f"  Saved to {csv_path}")
    
    # Create the plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # Get unique runs and colors
    unique_runs = sorted(df['run_id'].unique())
    colors = plt.cm.tab20(range(len(unique_runs)))
    
    # Plot French orientation (left panel)
    for i, run_id in enumerate(unique_runs):
        run_data = df[df['run_id'] == run_id].sort_values('prop')
        ax1.plot(run_data['prop'], run_data['fr_orientation'], 
                alpha=0.3, linewidth=1.5, color=colors[i % len(colors)])
    
    # Add median line
    medians = df.groupby('prop')['fr_orientation'].median()
    ax1.plot(medians.index, medians.values, 
            color='black', linewidth=4, label='Median', linestyle='--')
    
    # Add mean line
    means = df.groupby('prop')['fr_orientation'].mean()
    ax1.plot(means.index, means.values, 
            color='red', linewidth=3, label='Mean', linestyle=':')
    
    ax1.set_xlabel('French Proportion in Training', fontsize=12)
    ax1.set_ylabel('Language Orientation Score', fontsize=12)
    ax1.set_title(f'French Sentences (ablation={ablation_type})', fontsize=14, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim(0, 1)
    ax1.legend()
    
    # Plot Dutch orientation (right panel)
    for i, run_id in enumerate(unique_runs):
        run_data = df[df['run_id'] == run_id].sort_values('prop')
        ax2.plot(run_data['prop'], run_data['nl_orientation'], 
                alpha=0.3, linewidth=1.5, color=colors[i % len(colors)])
    
    # Add median line
    medians = df.groupby('prop')['nl_orientation'].median()
    ax2.plot(medians.index, medians.values, 
            color='black', linewidth=4, label='Median', linestyle='--')
    
    # Add mean line
    means = df.groupby('prop')['nl_orientation'].mean()
    ax2.plot(means.index, means.values, 
            color='red', linewidth=3, label='Mean', linestyle=':')
    
    ax2.set_xlabel('French Proportion in Training', fontsize=12)
    ax2.set_ylabel('Language Orientation Score', fontsize=12)
    ax2.set_title(f'Dutch Sentences (ablation={ablation_type})', fontsize=14, fontweight='bold')
    ax2.grid(True, alpha=0.3)
    ax2.set_ylim(0, 1)
    ax2.legend()
    
    # Overall title
    fig.suptitle(f'Language Orientation Score - Ablation: {ablation_type}', 
                fontsize=16, fontweight='bold', y=1.02)
    
    plt.tight_layout()
    
    # Save plot
    plot_path = plots_dir / 'orientation_line.png'
    plt.savefig(plot_path, dpi=150, bbox_inches='tight')
    plt.close()
    print(f"  Plot saved to {plot_path}")

print(f"\n{'='*70}")
print(f"All plots saved to {plots_base}")
print(f"{'='*70}")
```
